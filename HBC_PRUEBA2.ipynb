```python
import os
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support import expected_conditions as EC



options = Options()
options.add_argument("--disable-blink-features=AutomationControlled")  # Oculta Selenium
driver = webdriver.Chrome(service=service, options=options)
driver.quit()  # Cierra cualquier sesión previa antes de abrir una nueva
nombre_usuario = input("Ingresa el nombre del usuario de Instagram (sin @): ")

# Read credentials from file
with open('credenciales.txt', 'r') as f:
    us = f.readline().strip()
    pas = f.readline().strip()

# Use WebDriverManager to manage ChromeDriver
service = Service(ChromeDriverManager().install())

driver = webdriver.Chrome(service=service)

# Open Instagram
driver.get('https://www.instagram.com/')
# Locate the username field

# para cargar usuario de instagram 
time.sleep(5)
username = driver.find_element("name", "username")
username.click()
username.send_keys(us)
# para cargar contraseña de instagram 
time.sleep(5)
password_field = driver.find_element("name", "password")
password_field.click()
password_field.send_keys(pas)
time.sleep(5)
password_field.send_keys(Keys.RETURN)  # Simula presionar "Enter"
# Esperamos unos segundos después de iniciar sesión
time.sleep(5)
try: 
      time.sleep(5)
      info_se = driver.find_element(By.XPATH, "//div[@role='button' and text()='Ahora no']")
      info_se.click()
except Exception as e:
    print(f"No se pudo encontrar el botón de búsqueda o el campo de texto: {e}")

try:
    buscar_menu = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.XPATH, "//div[@class='x1iyjqo2 xh8yej3'and @data-visualcompletion='ignore-dynamic']")))
    # Usar JavaScript para modificar el color de fondo y el borde
    driver.execute_script("""
    arguments[0].style.backgroundColor = 'red';
    arguments[0].style.border = '3px solid blue';
    """, buscar_menu)
    print(f"Estoy dentro del menu de los iconos donde esta buscar")

    try:
        buscar_icono = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, "//span[@aria-describedby='«R56t5bb9klrj5ipd5aq»']//div[@class='x1n2onr6 x6s0dn4 x78zum5']//a[@href='#' and @role='link' and @tabindex='0']")))
        driver.execute_script("""
            arguments[0].style.backgroundColor = 'yellow';
            arguments[0].style.border = '3px solid blue';
            """, buscar_icono)
        buscar_icono.click()
        print("¡Se encontro el icono buscar")
        try:
            # Localizar el campo de búsqueda
            search_box = WebDriverWait(driver, 10).until(
             EC.element_to_be_clickable((By.XPATH, "//input[@aria-label='Buscar entrada']"))
            )
            
            search_box.click()
            
            search_box.send_keys(nombre_usuario)  
            time.sleep(2)  

            search_box.click()

            primer_resultado = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, "(//div[@class='x9f619 x78zum5 xdt5ytf x1iyjqo2 x6ikm8r x1odjw0f xh8yej3 xocp1fn']//a[@role='link'])[1]")))
            primer_resultado.click()

            print("Se esta realizando la busqueda de la persona")

            current_url = driver.current_url
            if(nombre_usuario in current_url):
                 print("Estamos en el perfil del usuario. Procediendo a extraer seguidores...")
                 time.sleep(3)
                 
                 modal = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.XPATH, "(//ul[@class='x78zum5 x1q0g3np xieb3on']//li[@class='xl565be x1m39q7l x1uw6ca5 x2pgyrj'])[2]"))
                )
                 modal.click()

                 for _ in range(10):  # Puedes cambiar 10 por otro número si necesitas más scroll
                    driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", modal)
                    time.sleep(1.5)
                 # Aquí va el resto del código: abrir seguidores, hacer scroll, extraer y guardar
            else:
                print("No se ha ingresado correctamente al perfil del usuario.")
                
            

        except Exception as e:
            print(f"No selecciono la primera opcion: {e}")

    except Exception as e:
        print(f"No entro al boton buscar")

except Exception as e:
    print(f"No se encuentra en el menu donde se enceuntra varias opciones")
```
